ある数字の集合があったときに、それが正規分布に従うと仮定して、その正規分布に従う乱数をたくさん得たい。

これを実行する関数を作ってみました。

```python
import numpy as np
def random_gauss(numbers, size=10):
    return np.random.normal(loc = np.mean(numbers), scale = np.std(numbers), size  = size)
```

簡単、簡単。これで、たとえば得られた数字が [5, 4, 5, 6, 5, 4, 5] で、それが正規分布に従うと仮定して、その正規分布に従う乱数を 10 個得たいとします。

```python
random_gauss([5, 4, 5, 6, 5, 4, 5], 10)
```

10x10 の配列にしたければ、このように使えます。

```python
random_gauss([5, 4, 5, 6, 5, 4, 5], (10, 10))
```
